```python
#Convert time to datetime
```

```python

data['time']=pd.to_datetime(data['time'])
data.info()
```

```python
 #Drop records with missing latitude, longitude, mag
```

```python

print(data.isnull().sum())
```

```python
data = data.dropna(subset=['latitude','longitude','mag'])
```

```python
# Filter only earthquakes with magnitude >= 4.0
data[data['mag'] >= 4.0]

```

```python
#Add a column day_of_week from time
data['day_of_week'] =data['time'].dt.day_name()
data.columns
```

Create a column severity_level based on magnitude:

○ <4.0: "Low"

○ 4.0 - 6.0: "Moderate"

○ 6.0+: "High"

```python
conditions =[
    (data['mag']<4.0),
    (data['mag']>=4.0)&(data['mag']<=6.0),
    (data['mag']>6.0)
]
choices=['Low','Moderate','High']
data['severity_level']=np.select(conditions,choices,default='unknown')
```

```python
data.columns
```

```python
#count number of earthquakes per place

earthquake_count=data['place'].value_counts()
print(earthquake_count)
```

```python
#Compute average magnitude and max depth per day
daily_stats = data.groupby('day_of_week').agg(
    average_magnitude=('mag', 'mean'),
    max_depth=('depth', 'max')
)
display(daily_stats)
```

```python
data.to_csv('cleaned_earthquakes.csv')
daily_stats.to_csv('earthquake_summary.csv')
```
